Generates good lookin graphs from metrics output files

```python
# Import packages
from __future__ import division, print_function # Imports from __future__ since we're running Python 2
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%run '/home/mark/Desktop/Project/Metrics-Analysis/Utilities.ipynb'
```

Program parameters:

```python
folder_path1            = '../../Testing/Plastic-Parallel-Programming/build/runs/Experiment 1/run1'
folder_path2            = '../../Testing/Plastic-Parallel-Programming/build/runs/Experiment 1/run2'
rootword               = 'Experiment'
numOfExperiments       = 1
numOfRepeats           = 50
```

Set files to read

```python
file_names = []

for i in range(numOfExperiments):
    repeat_file_names = []
    
    for j in range(numOfRepeats):
        repeat_file_names.append(rootword + str(i + 1) + "_Repeat" + str(j) + ".csv")
        
    file_names.append(repeat_file_names)
```

Import the dataset

```python
data1 = []
data2 = []
        
fetch_data(folder_path1, file_names, data1)
fetch_data(folder_path2, file_names, data2)
        
data1 = extract_runtimes(data1)
data2 = extract_runtimes(data2)

gen_data1 = generate_data(data1, [1, 2, 3, 4], 'No Metrics Collection')
gen_data2 = generate_data(data2, [1, 2, 3, 4], 'Metrics Collection')

dataset = pd.DataFrame(gen_data1 + gen_data2, 
                       columns=['Number of Cores', 'Metrics', 'Runtime (Milliseconds)'])

np.average(data2[0])
```

```python
sns.set(style="whitegrid", font_scale=1.75)

graph = sns.factorplot(x="Metrics", y = "Runtime (Milliseconds)", data=dataset, 
                   size=10, kind="bar", palette="muted")

graph.axes[0,0].set_ylim(20000,)

graph.despine(left=True)
graph.savefig("Graphs/experiment1.png")

sns.plt.show()
```
